# Color Detection from Image

```python
import cv2
import pandas as pd

# image ni load chesi
img_path = r"C:\Users\Yaswa\Downloads\multi_color.jpg"
img = cv2.imread(img_path)

if img is None:
    print("Error: Image not found. Check the file path.")
    exit()

# output screen correctyga ravali so, andhuke resize chesi
screen_width = 800
screen_height = 600
img = cv2.resize(img, (screen_width, screen_height))

# download chesukuna dataset ni load chesukundham
index = ["color", "color_name", "hex", "R", "G", "B"]
csv_path = r"C:\Users\Yaswa\Downloads\Image_Detection\colors.csv"
try:
    data = pd.read_csv(csv_path, names=index)
except Exception as e:
    print(f"Error reading CSV: {e}")
    exit()

# click cheyakunda answer vasthe bagodhuga
clicked = False
r = g = b = x_pos = y_pos = 0  # so, ekkada color values in and positions ni initialize chesukundham

# Mouse click function
def draw_function(event, x, y, flags, param):
    global clicked, r, g, b, x_pos, y_pos
    if event == cv2.EVENT_LBUTTONDBLCLK:  # Left Double Click
        clicked = True
        x_pos, y_pos = x, y  # Store clicked position
        b, g, r = img[y, x]  # Get BGR color from image
        b, g, r = int(b), int(g), int(r)

# Function use chesi cloest color name esthundhi
def getColor(R, G, B):
    min_distance = float('inf')
    cname = "Unknown"
    for i in range(len(data)):
        d = abs(R - int(data.loc[i, "R"])) + abs(G - int(data.loc[i, "G"])) + abs(B - int(data.loc[i, "B"]))
        if d < min_distance:
            min_distance = d
            cname = data.loc[i, "color_name"]
    return cname

# OpenCV window
cv2.namedWindow("Color Detection", cv2.WINDOW_AUTOSIZE)
cv2.setMouseCallback("Color Detection", draw_function)

while True:
    img_copy = img.copy()  # Use a copy to prevent permanent drawing

    if clicked:
        # text ki background kosam oka rectangle box
        cv2.rectangle(img_copy, (10, 20), (600, 60), (b, g, r), -1)
        text = getColor(r, g, b) + f' ({r},{g},{b})'

        # output screen lo text kosam
        text_color = (0, 0, 0) if (r + g + b) >= 400 else (255, 255, 255)
        cv2.putText(img_copy, text, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 0.8, text_color, 2)

    cv2.imshow("Color Detection", img_copy)

    # Break loop if 'Esc' key is pressed
    if cv2.waitKey(1) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
